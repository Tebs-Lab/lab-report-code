```python
from matplotlib import pyplot as plt
import numpy as np

# evenly sampled time at 200ms intervals
x = np.arange(0., 50, 0.1)
y = 2*x + 400/x


plt.title("Zoomed Out")
plt.plot(x, y)
```

```python
x_zoom = np.arange(12, 16, .01)
y_zoom = 2*x_zoom + 400/x_zoom

plt.title("Zoomed In To Inflection Point")
plt.plot(x_zoom, y_zoom)
plt.scatter(14.14, 2*14.14 + 400 / 14.14, color='r')
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(12,6))

ax1.title.set_text("Zoomed Out")
ax1.plot( x, y)

ax2.title.set_text("Zoomed In To Inflection Point")
ax2.plot( x_zoom, y_zoom)
ax2.scatter(14.14, 2*14.14 + 400 / 14.14, color='r')
```
